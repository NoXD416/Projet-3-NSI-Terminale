## Compte à rebours

```python
def saisir_temps_secondes():
    secondes = int(input("Entrer le nombre de secondes : "))
    minutes = int(input("Entrer le nombre de minutes: "))
    heurs = int(input("Entrer le nombre d'heurs: "))
    return secondes + minutes*60 + heurs*3600

def afficher_temps(temps):
    background("black")
    fill("lightblue")
    text(temps%60, 635, 225) 
    text((temps//60) % 60, 335, 225)
    text(temps//3600, 50, 225)
    
    fill("red")
    circle(245, 225, 40)
    circle(245, 165, 40)
    circle(550, 225, 40)
    circle(550, 165, 40)

def setup():
    createCanvas(800,400)
    frameRate(1)
    textSize(100)

def draw():
    global entree
    afficher_temps(entree)
    entree = entree - 1
    if entree == -1:
        background("black")
        fill("orange")
        text("C'est terminé !", 50, 225)
        stop()
    
entree = saisir_temps_secondes()
run()
```

## Horloge binaire version 1

```python
def binaire(entier, nb_carreaux):
    entier = int(entier)
    binaire = [0]*nb_carreaux
    compteur = 1
    while entier != 0:
        binaire[nb_carreaux-compteur] = entier%2
        entier = entier // 2
        compteur = compteur + 1
    return binaire


def dessiner_colonne(nb_binaire, position, hauteur):
    for bit in nb_binaire:
        if bit == 1:
            fill(31, 81, 255)
            square(position, hauteur, 100)
        else:
            fill(172, 173, 172)
            square(position, hauteur, 100)
        hauteur = hauteur +  100
    
    
def fonction_principale(temps, position, hauteur):
    temps = str(temps)
    if len(temps) == 2:
        dessiner_colonne(binaire(temps[0], 4), position, hauteur)
        dessiner_colonne(binaire(temps[1], 4), position+100, hauteur)
    
    else:
        dessiner_colonne([0,0,0,0], position, hauteur)
        dessiner_colonne(binaire(temps[0], 4), position+100, hauteur)

def setup():
    createCanvas(1000, 600)    


def draw():
    background(26, 17, 16)
    fonction_principale(hour(), 100, 100)
    fonction_principale(minute(), 400, 100)
    fonction_principale(second(), 700, 100)

run()
```

## Horloge Binaire version 2

```python
def binaire(entier, nb_carreaux):
    entier = int(entier)
    binaire = []
    while entier != 0:
        binaire.append(entier%2)
        entier = entier // 2
    return binaire


def dessiner_colonne(nb_binaire, position, hauteur, nb_carreaux):
    
    for bit in nb_binaire:
        
        if bit == 1:
            fill(31, 81, 255)
            square(position, hauteur, 100)
        
        else:
            fill(172, 173, 172)
            square(position, hauteur, 100)
        
        hauteur = hauteur - 100
        nb_carreaux = nb_carreaux - 1
    
    if nb_carreaux > 0:
        
        for _ in range(nb_carreaux):
            fill(172, 173, 172)
            square(position, hauteur, 100)
            hauteur = hauteur - 100
            
    
    
def fonction_principale(temps, position, hauteur, nb_carreaux_col1, nb_carreaux_col2):
    temps = str(temps)
    if len(temps) == 2:
        dessiner_colonne(binaire(temps[0], 4), position, hauteur, nb_carreaux_col1)
        dessiner_colonne(binaire(temps[1], 4), position+100, hauteur, nb_carreaux_col2)
    
    else:
        dessiner_colonne([0], position, hauteur, nb_carreaux_col1)
        dessiner_colonne(binaire(temps[0], 4), position+100, hauteur, nb_carreaux_col2)

def setup():
    createCanvas(1000, 600)    


def draw():
    background(26, 17, 16)
    fonction_principale(hour(), 100, 400, 2, 4)
    fonction_principale(minute(), 400, 400, 3, 4)
    fonction_principale(second(), 700, 400, 3, 4)

run()
```

## Horloge Fibonnaci

```python
def prelever_temps():
    heurs = hour()
    if heurs <= 12:
        return heurs, minute()//5
    return heurs%12, minute()//5

def combinaison_fibonnaci(temps):
    fibonnaci, dico_fibo, indice = [5,3,2,1,1], {"5": 0, "3": 0, "2": 0, "1": 0, "1'": 0}, 0
    
    while temps > 0:
        if fibonnaci[indice] <= temps:
            dico_fibo[str(fibonnaci[indice])] += 1
            temps -= fibonnaci[indice]
        indice += 1
    
    if dico_fibo["1"] == 2:
        dico_fibo["1'"], dico_fibo["1"] = 1, 1
       
    return dico_fibo

def fonction_principale():
    heurs, minutes = prelever_temps()
    carreaux = {"5": [300,0,500], "3": [0,200,300] , "2": [0,0,200] , "1":[200,0,100] , "1'": [200,100,100]}
    dico_heurs, dico_minutes = combinaison_fibonnaci(heurs), combinaison_fibonnaci(minutes)
    
    for cle in dico_heurs:
        
        if dico_heurs[cle] == 0 and dico_minutes[cle] == 0:
            fill(211, 211, 211)
        
        elif dico_heurs[cle] == 1 and dico_minutes[cle] == 1:
            fill(77, 77, 255)
            
        elif dico_heurs[cle] == 1 and dico_minutes[cle] == 0:
            fill(255, 49, 49)
        
        elif dico_heurs[cle] == 0 and dico_minutes[cle] == 1:
            fill(57, 255, 20)
        
        square(carreaux[cle][0],carreaux[cle][1],carreaux[cle][2])

def setup():
    createCanvas(800, 500)
    frameRate(1)
    
def draw():
    background(255,255,255)
    fonction_principale()
    
run()
```
